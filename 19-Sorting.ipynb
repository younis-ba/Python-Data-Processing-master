```python
import pandas as pd
import numpy as np
```

```python
# here we create a simple dataframe with random values
raw_data = {
    'col1' : np.random.normal(size=5),
    'col2' : np.random.normal(size=5),
    'col3' : np.random.normal(size=5),
    'col4' : np.random.normal(size=5)
}
my_df = pd.DataFrame(raw_data, index = ['r1', 'r2', 'r3', 'r4', 'r5'])
my_df
```

```python
my_df.sort_values(by='col1')
```

```python
my_df.sort_values(by=['col1','col4'])
```
